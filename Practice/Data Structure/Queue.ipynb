# Queue
- FIFO(First In First Out)
- LILO(Last In Last Out)
- Two Ends
  - Head
  - Tail

### Functions
- Enqueue
    - Adding Element
- Dequeue
    - Removing Element
- IsFull
    - Checks queue is full or not 
- IsEmpty
    - Checks queue is empty or not 

## Using List

```python
Queue = []
```

```python
Capacity = 10
def Enqueue(Value):
    if(IsFull() == False):
        Queue.append(Value)
    else:
        print("Empty List")
    

def Dequeue():
    if(IsEmpty() == False):
        Queue.pop(0)
    else:
        print("Dequeue from empty list")

def IsFull():
    if(len(Queue) == 10):
        return True
    else:
        return False
    
def IsEmpty():
    return (not Queue)
    
def FirstValue():
    if(IsEmpty() == False):
        print(Queue[0])
    else:
        print("Empty List")

def LastValue():
    if(IsEmpty() == False):
        print(Queue[-1])
    else:
        print("Empty List")
```
