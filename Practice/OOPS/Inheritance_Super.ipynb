```python
class Item:
    def __init__(self,name: str,price: int,quantity = 1): 
        self.name = name
        self.price = int(price)
        self.quantity = int(quantity)

    def Calculate(self):
        print("Total Price : ",self.price * self.quantity)
```

# Without Super

```python
class Phone(Item):
    def __init__(self,name: str,price: int,quantity = 1,num = 0):
        self.name = name
        self.price = int(price)
        self.quantity = int(quantity)  # without name,price,quantity , we can't use the parent class functions
        
        self.num = num
        print(self.num)
    
```

```python
I1 = Item("Phone",500)
I1.Calculate()
```

```python
P1 = Phone("Phone",500,2,5)
P1.Calculate()
```

# Super

```python
class Phone(Item):
    def __init__(self,name: str,price: int,quantity = 1,num = 0):
        super().__init__(name,price,quantity)
        
        self.num = num
        print(self.num)
```

```python
P1 = Phone("Phone",500,2,5)
P1.Calculate()
```

# Without any attributes of parent

```python
class Phone(Item):
    def __init__(self,num = 0):
        self.num = num
        print(self.num)
```

```python
P1 = Phone("10")
```

# List OF Objects

```python
class Item:
    AllObject = []
    
    def __init__(self,name: str,price: int,quantity = 1): 
        
        self.name = name
        self.price = int(price)
        self.quantity = int(quantity)

        Item.AllObject.append(self)
    def __repr__(self):
        return f"Item({self.name},{self.price},{self.quantity})"

I1 = Item("Phone",500,3)
I2 = Item("Lap",50,3)
I3 = Item("Watch",5)

print(Item.AllObject)
```

```python
class Phone(Item):
    AllObject = []
    def __init__(self,name: str,price: int,quantity = 1,num = 0):
        super().__init__(name,price,quantity)
        
        self.num = num

        Phone.AllObject.append(self)

Phone("1",234)
Phone("3",34)
Phone("2",55,2,4)

print(Phone.AllObject)
print(Item.AllObject)
```

# To avoid This


```python
class Item:
    AllObject = []
    
    def __init__(self,name: str,price: int,quantity = 1): 
        
        self.name = name
        self.price = int(price)
        self.quantity = int(quantity)

        Item.AllObject.append(self)
    def __repr__(self):
        return f"{self.__class__.__name__}({self.name},{self.price},{self.quantity})"  # "Item" -> self.__class__.__name__

I1 = Item("Phone",500,3)
I2 = Item("Lap",50,3)
I3 = Item("Watch",5)

class Phone(Item):
    # AllObject = []
    def __init__(self,name: str,price: int,quantity = 1,num = 0):
        super().__init__(name,price,quantity)
        
        self.num = num

        # Phone.AllObject.append(self) because We are calling a super Function

Phone("1",234)
Phone("3",34)
Phone("2",55,2,4)

print(Phone.AllObject)  # It Calls the AllObject From the parent class
print(Item.AllObject)
```
