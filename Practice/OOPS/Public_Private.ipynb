# Partially Public

**Error**


```python
class Item:
    def __init__(self,name: str,price: int,quantity = 1): 
        self._name = name   # _ to access the name to change
        self.price = int(price)
        self.quantity = int(quantity)

    @property
    def name(self):
        return self._name
```

```python
I1 = Item("Ph",3)
```

**While Changing**


```python
I1._name = "Phone"
```

```python
print(I1.name)
```

```python
I1.__dict__
```

# Private

```python
class Item:
    def __init__(self,name: str,price: int,quantity = 1): 
        self.__name = name   # __ to Private
        self.price = int(price)
        self.quantity = int(quantity)

    @property
    def name(self):
        return self.__name
```

```python
I1 = Item("Ph",3)
```

```python
I1.name
```

**Can't Change After Created Object**


```python
I1 = Item("Ph",3)
```
