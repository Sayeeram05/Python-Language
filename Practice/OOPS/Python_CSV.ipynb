```python
class Item:
    AllObjects = []
    
    def __init__(self,name: str,price: int,quantity = 1): 
        self.name = name
        self.price = int(price)
        self.quantity = int(quantity)
        
        Item.AllObjects.append(self)
    def __repr__(self):
        return f"Item({self.name}, {self.price}, {self.quantity})"
    
    @classmethod
    def Data_From_csv(cls):
        with open("Data.csv", "r") as D:
            Dict = csv.DictReader(D)
            AllData = list(Dict)
        for i in AllData:
            Item(i["Name"],i["price"],i["Qty"])
```

```python
Item.Data_From_csv()
```

```python
Item.AllObjects
```

```python
for i in Item.AllObjects:
    print(i)
```

```python
for i in Item.AllObjects:
    print(i.name,i.price,i.quantity)
```

# Static Method

```python
class Item:
    @staticmethod
    def StaticMethod1():
        print("This Is Like Normal Function")
    def CheckInt(num):
        if(isinstance(num, float)):
            return num.is_integer() # num.is_integer() -> for 5.0,10.0 to consider as int
        elif(isinstance(num , int)):
            return True
        else:
            return False        
```

```python
Item.StaticMethod()
```

```python
print(Item.CheckInt(8.0))
print(5.0.is_integer())
print(5.1.is_integer())
```
